```ruby
require 'daru'
require 'statsample-glm'
require 'stopwords'
require 'lingua/stemmer'
require 'numo/narray'
require 'nmatrix'
```

```ruby
pos = File.open("../data/rt-polaritydata/rt-polarity.pos").read
neg = File.open("../data/rt-polaritydata/rt-polarity.pos").read

pos_tagged = pos.split("\n").map {|line| ["+1", line]}
neg_tagged = neg.split("\n").map {|line| ["-1", line]}
sen = (pos_tagged + neg_tagged).shuffle
counter = {"+1" => 0, "-1" => 0}

sen.each do |line|
  counter[line[0]] += 1
end

counter
```

```ruby
# in order to check if the word is stop word with O(1)
Stop_words = Hash.new {|hash, key| hash[key] = false}
(Stopwords::STOP_WORDS + ["'", ",", '.', ":", ";"]).each {|stop_word| Stop_words[stop_word] = true}

def is_stop?(word)
  Stop_words[word]
end
```

```ruby
all_vocab = []
features = []
labels = []

sen.each_with_index do |(pos, text), index|
  tokens = text.split.map{|token| Lingua.stemmer(token).downcase!}.reject{|token| is_stop?(token)}
  all_vocab += tokens
  counter = Hash.new{|hash, key| hash[key] = 0}
  tokens.each{ |token| counter[token] += 1}
  labels << (pos == "+1" ? 1 : 0)
  features << counter
end

all_vocab.uniq!

token2id = {}
id2token = {}

all_vocab.each_with_index do |token, id|
  token2id[token] = id
  id2token[id] = token
end

1
```
